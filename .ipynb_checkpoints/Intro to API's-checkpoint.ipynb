```python
my_racer_data = response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']
my_racer_data
```

```python
def get_racer_info(data):
    new_data = []
    for racer in data:
        racer_dict = {}
        racer_name = f"{racer['Driver']['givenName']} {racer['Driver']['familyName']}"
        
        racer_dict[racer_name]={
            'first_name':racer['Driver']['givenName'],
            'last_name':racer['Driver']['familyName'],
            'position':racer['position'],
            'wins':racer['wins'],
            'DOB':racer['Driver']['dateOfBirth'],
            'nationality':racer['Driver']['nationality'],
            'constructor':racer['Constructors'][0]['name']
        }
        new_data.append(racer_dict)
    return new_data
    
get_racer_info(my_racer_data)
```

```python
def get_driver_info_by_year_and_round(year, r):
    url = f"https://ergast.com/api/f1/{year}/{r}/driverStandings.json"
    reponse = requests.get(url)
    if not response.json()['MRData']['StandingsTable']['StandingsLists']:
        return 'We had an error loading your data. It is likely the year or round is not in the database.'
    data = response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']
    return get_racer_info(data)
    
get_driver_info_by_year_and_round(2019, 4)
```

```python
import random as rd
from collections import defaultdict

# function to get pokemon's name
def get_pokemon_name(pokemon_id):
    url = f"https://pokeapi.co/api/v2/pokemon/{str(pokemon_id)}/"
    response = requests.get(url)
    data = response.json()
    return data["name"]


# function to get pokemon's type, abilities, weight info
def get_pokemon_info(pokemon_name):
    url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_name}/"
    response = requests.get(url)
    data = response.json()
    pokemon_type = data['types'][0]['type']['name']
    abilities = data['abilities']
    ability_ls = []
    for i in abilities:
        item = i['ability']['name']
        ability_ls.append(item)

    pokemon_dict = {}
    pokemon_dict[pokemon_type.title()]= {
        data['name']: {
            'abilities': ability_ls,
            'weight': data['weight']
        }
    }
    return pokemon_dict


# function to create 20 pokemons (no selection)
def create_20pokemons():
    pokemon_dict = defaultdict(dict)
    for i in range(1,21):
        info = get_pokemon_info(get_pokemon_name(i))
        for key, value in info.items():
            for name, data in value.items():
                pokemon_dict[key][name] = data
    pokemon_dict = (dict(pokemon_dict))
    return pokemon_dict


# function to create pokemons (with selection)
def create_pokemons():
    print("\nYour pokemon options:\n")

    count = 0
    for x in range(50):
        data = my_pokemon_data['results'][x]['name']
        count += 1
        print(f"{count}= {data}", end="\t\t")
        
    try:

        my_pokemon_list = []
        prompt = "\n\n Enter the id number of the pokemon you want to create or 'quit': "
        selection = input(prompt)

        pok_ct = 0
        while selection.lower() != 'quit':
            if len(my_pokemon_list) < 20:

                my_pokemon_list.append(get_pokemon_info(get_pokemon_name(selection)))
                dd = defaultdict(dict)
                for d in my_pokemon_list:
                    for key, value in d.items():
                        dd[key].update(value)
                final_ls = [dict(dd)]
                pok_ct += 1
                print(f"({pok_ct}/20) Creating: {get_pokemon_name(selection)}.....")
            else:
                print("\n\n------- These are the 20 pokemons you created -------\n")
                return final_ls
                break


            selection = input(prompt)
        else:
            print("See you later!")
            
    except:
        print("\nOops, something went wrong. Please try again.")
        print("(Please enter pokemon ID# 1-50 or 'quit' to quit.)")
```

```python
# create 20 pokemons (no selection)
print("\n------- These are the 20 pokemons you created: -------\n")
create_20pokemons()
```
